```python
def randomQuestions(category):
    questions_asked_sector = ['Food and drink', 'Clothing and footwear', 'Construction', 'Cheers', 'Education', 'Services', 'IT and technology', 'Sales and Marketing']
    questions_asked_business_type = ['Micro enterprise', 'Microentrepreneur', 'Small business', 'Large-sized company', 'Does not have_sector']
    questions_asked_interests = ['Learn about a specific topic', 'Meet other people', 'Present my business', 'Search for new opportunities']
    sector = random.choice(questions_asked_sector)
    business_type = random.choice(questions_asked_business_type)
    interests = random.choice(questions_asked_interests)
    return [sector,business_type, interests]
```

```python
df[['sector', 'business_type', 'interests']] = df.apply(lambda x: pd.Series(randomQuestions(x.category)), axis=1)
```
